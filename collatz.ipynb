```python
import matplotlib.pyplot as plt
```

```python
n = int(input("Enter initial number: "))
i = 1;
n;
nmax = n;
arr =[n]
```

```python
while n > 1:
    i = i+1;
    if n%2 == 1:
        n = 3*n +1;
    else:
        n = n / 2;
    if n > nmax:
        nmax = n;
    arr.append(n)
```

```python
plt.plot(arr, marker="D")
plt.title("Collatz Conjecture")
plt.xlabel("Iteration")
plt.ylabel("Number")
print("The number of iteration is: ", i)
print("The highest number is: ", nmax)
```
